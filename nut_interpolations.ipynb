## Water Temperature Interpolations

```python
#import datasets to be filled for each point of interest
#all updated to 6/30/22

#nf above confluence
nf_above_conf_temp = pd.read_csv(r'/Users/maanavjhatakia/Documents/Python.nosync/geosmin/nf_above_conf_water_temp.csv')

#ms above confluence
ms_above_conf_temp = pd.read_csv(r'/Users/maanavjhatakia/Documents/Python.nosync/geosmin/ms_above_conf_water_temp.csv')

#upstream strontia
upstream_strontia_temp = pd.read_csv(r'/Users/maanavjhatakia/Documents/Python.nosync/geosmin/up_strontia_water_temp.csv')

#downstream strontia
downstream_strontia_temp = pd.read_csv(r'/Users/maanavjhatakia/Documents/Python.nosync/geosmin/ds_strontia_water_temp.csv')

#sp below chatfield
sp_below_chat_temp = pd.read_csv(r'/Users/maanavjhatakia/Documents/Python.nosync/geosmin/sp_bel_chat_water_temp.csv')
```

```python
#fill in missing data with a polynomial trend of order 2 between 1st of each month
nf_above_conf_temp = nf_above_conf_temp.interpolate(method='polynomial', order = 2)
ms_above_conf_temp = ms_above_conf_temp.interpolate(method='polynomial', order =2)
upstream_strontia_temp = upstream_strontia_temp.interpolate(method = 'polynomial', order = 2)
downstream_strontia_temp = downstream_strontia_temp.interpolate(method = 'polynomial', order = 2)
sp_below_chat_temp = sp_below_chat_temp.interpolate(method = 'polynomial',order = 2)
```

```python
#save dataframe to csv file to ease merging
nf_above_conf_temp.to_csv('nf_above_conf_temp_filled.csv', index = False)
ms_above_conf_temp.to_csv('ms_above_conf_temp_filled.csv', index = False)
upstream_strontia_temp.to_csv('upstream_strontia_temp_filled.csv', index = False)
downstream_strontia_temp.to_csv('downstream_strontia_temp_filled.csv', index = False)
sp_below_chat_temp.to_csv('sp_below_chat_temp_filled.csv', index = False)
```
