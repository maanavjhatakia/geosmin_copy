```python
#import datasets to be filled for each point of interest
#all updated to 6/30/22

#nf above confluence
nf_above_conf_temp = pd.read_csv(r'C:\Users\MJHATA\OneDrive - Denver Water\Documents\Python Scripts\nf_above_conf_water_temp.csv')

#ms above confluence
ms_above_conf_temp = pd.read_csv(r'C:\Users\MJHATA\OneDrive - Denver Water\Documents\Python Scripts\ms_above_conf_water_temp.csv')

#upstream strontia
upstream_strontia_temp = pd.read_csv(r'C:\Users\MJHATA\OneDrive - Denver Water\Documents\Python Scripts\up_strontia_water_temp.csv')

#downstream strontia
downstream_strontia_temp = pd.read_csv(r'C:\Users\MJHATA\OneDrive - Denver Water\Documents\Python Scripts\ds_strontia_water_temp.csv')

#sp below chatfield
sp_below_chat_temp = pd.read_csv(r'C:\Users\MJHATA\OneDrive - Denver Water\Documents\Python Scripts\sp_bel_chat_water_temp.csv')
```

```python
#fill in missing data with a polynomial trend of order 2 between 1st of each month
nf_above_conf_temp = nf_above_conf_temp.interpolate(method='polynomial', order = 2)
ms_above_conf_temp = ms_above_conf_temp.interpolate(method='polynomial', order =2)
upstream_strontia_temp = upstream_strontia_temp.interpolate(method = 'polynomial', order = 2)
downstream_strontia_temp = downstream_strontia_temp.interpolate(method = 'polynomial', order = 2)
sp_below_chat_temp = sp_below_chat_temp.interpolate(method = 'polynomial',order = 2)
```
